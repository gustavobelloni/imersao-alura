Desafios aula 2: 

1 - Criar uma escala de valores em milhões.

2 - Deixar o gráfico do histograma de valores legível (alterar os labels, cores, título, escala).

3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal.

4 - Pesquisar uma visualização para analisar os quartis, mediana e outliers. 

```python
# 1 - Criar uma escala de valores em milhões.
dados_vendas['Valor_milhoes'] = dados_vendas['Valor_anuncio_float']/1000000
dados_vendas.describe()
```

```python
# 2 - Deixar o gráfico do histograma de valores legível (alterar os labels, cores, título, escala).

sns.set()
plt.figure(figsize=(10,8))
ax = sns.histplot(data = dados_vendas, x='Valor_milhoes', bins=250, kde=True, color='#1B7001')
ax.set_title('Histograma Valor Imóvel')
ax.set_xlabel('Valor dos Imóveis em Milhões')
ax.set_ylabel('Nº de Imóveis')
plt.xlim(-0.05, 10.2)
plt.xticks(list(range(0,11)))
plt.show()
```

```python
# 3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal
# Desafio feito com auxílio da aula 3
# Calculando o preço do metro quadrado por bairro
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem']
dados_vendas.groupby('Bairro').mean() # Transforma o bairro em index e pega apenas as colunas de valores possíveis de se usar média 
```

```python
dados_bairro = dados_vendas.groupby('Bairro').sum()
dados_bairro
```

```python
dados_bairro['Valor_m2_bairro'] = dados_bairro['Valor_anuncio_float']/dados_bairro['Metragem']
dados_bairro
```

```python
# Pegando os valores dos 10 primeiros bairros com os maiores valores
top_bairros = dados_vendas['Bairro'].value_counts()[:10].index
```

```python
# Resetando o grafico para o bairro não ser lido como index
dados_bairro.reset_index(inplace = True)
dados_bairro
```

```python
# Criando uma query com os top 10 bairros
dados_bairro.query('Bairro in @top_bairros')
```

```python
# Criando gráfico de barra sobre o valor de m2 quadrado por bairro
plt.figure(figsize=(10,8))
ax = sns.barplot(x='Bairro', y='Valor_m2_bairro', data=dados_bairro.query('Bairro in @top_bairros'))
ax.tick_params(axis='x', rotation=45)
```

```python
# 4 - Pesquisar uma visualização para analisar os quartis, mediana e outliers.
# Criando gráfico de boxplot sobre o valor de anúncio por bairro para analisar
plt.figure(figsize=(10,8))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @top_bairros'), x='Bairro', y='Valor_anuncio_float')
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Criando gráfico de boxplot sobre o valor da metragem por bairro eliminando os outliers
plt.figure(figsize=(10,8))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @top_bairros and Metragem < 30000'), x='Bairro', y='Metragem')
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Criando gráfico de boxplot para análisar os quartis, mediana e outliers das vagas.
plt.figure(figsize=(8, 12))
ax = sns.boxplot(y = 'Vagas',data = dados_vendas, color='green')
ax.set_title("Distribuicao das Vagas")
ax.set_xlabel("Vagas")
ax.set_ylabel("Numero de Vagas")
plt.show()
```
